```python
import spacy
import joblib
import pandas as pd
nlp=spacy.load('en_core_web_sm')
```

```python
def rule1(text): 
    doc = nlp(text) 
    sent = []
    for token in doc: 
        if (token.pos_=='VERB'):
            phrase = "" 
            for sub_tok in token.lefts: 
                if (sub_tok.dep_ in ['nsubj','nsubjpass']) and (sub_tok.pos_ in ['NOUN','PROPN','PRON'] ):
                    phrase += sub_tok.text 
                    phrase += " "+token.text
                    for sub_tok in token.rights:
                        if (sub_tok.dep_ in ['dobj']) and (sub_tok.pos_ in ['NOUN','PROPN'] ):
                            phrase += " "+sub_tok.text
                            sent.append(phrase)
    return sent
```

```python
def rule2(text):
    
    doc = nlp(text)

    pat = []
    
    # iterate over tokens
    for token in doc:
        phrase = ''
        # if the word is a subject noun or an object noun
        if (token.pos_ == 'NOUN')\
            and (token.dep_ in ['dobj','pobj','nsubj','nsubjpass']):
            
            # iterate over the children nodes
            for subtoken in token.children:
                # if word is an adjective or has a compound dependency
                if (subtoken.pos_ == 'ADJ') or (subtoken.dep_ == 'compound'):
                    phrase += subtoken.text + ' '
                    
            if len(phrase)!=0:
                phrase += token.text
             
        if  len(phrase)!=0:
            pat.append(phrase)
        
    
    return pat
```

```python
def rule0(text, index):
    
    doc = nlp(text)
        
    token = doc[index]
    
    entity = ''
    
    for sub_tok in token.children:
        if (sub_tok.dep_ in ['compound','amod']):# and (sub_tok.pos_ in ['NOUN','PROPN']):
            entity += sub_tok.text+' '
    
    entity += token.text

    return entity
```

```python
def rule3(text):
    
    doc = nlp(text)
    
    sent = []
    
    for token in doc:

        if token.pos_=='ADP':

            phrase = ''
            if token.head.pos_=='NOUN':
                
                # appended rule
                append = rule0(text, token.head.i)
                if len(append)!=0:
                    phrase += append
                else:  
                    phrase += token.head.text
                phrase += ' '+token.text

                for right_tok in token.rights:
                    if (right_tok.pos_ in ['NOUN','PROPN']):
                        
                        right_phrase = ''
                        # appended rule
                        append = rule0(text, right_tok.i)
                        if len(append)!=0:
                            right_phrase += ' '+append
                        else:
                            right_phrase += ' '+right_tok.text
                            
                        phrase += right_phrase
                
                if len(phrase)>2:
                    sent.append(phrase)
                

    return sent
```

```python
def rule4(text):
    
    doc = nlp(text)
    
    sent = []
    
    for token in doc:
        
        # If the token is a verb
        if (token.pos_=='AUX'):
            
            phrase =''
            
            # Only extract noun or pronoun subjects
            for sub_tok in token.lefts:
                
                if  (sub_tok.pos_ in ['NOUN','PROPN','PRON']):
                    
                    # Add subject to the phrase
                    phrase += sub_tok.text

                    # Save the root of the word in phrase
                    phrase += ' '+token.lemma_ 

                    # Check for noun or pronoun direct objects
                    for sub_tok in token.rights:
                        
                        # Save the object in the phrase
                        if (sub_tok.pos_ in ['NOUN','PROPN']):
                                    
                            phrase += ' '+sub_tok.text
                            sent.append(phrase)
            
    return sent
```

```python
def output(rule, text):
    output_bok = []

    for i in text:
        output_bok.append(rule(i))

    df_rule_bok = pd.DataFrame({'sentences':text,'output':output_bok})
    return df_rule_bok
```

```python
def dataframe1(df_rule):
    verb_dict = dict()
    dis_dict = dict()
    dis_list = []

    # iterating over all the sentences
    for i in range(len(df_rule)):

        # sentence containing the output
        sentence = df_rule.loc[i,'sentences']
        # output of the sentence
        output = df_rule.loc[i,'output']

        # iterating over all the outputs from the sentence
        for sent in output:

            # separate subject, verb and object
            n1 = sent.split()[:1]
            v = sent.split()[1]
            n2 = sent.split()[2:]

            # append to list, along with the sentence
            dis_dict = {'sentence':sentence,'Noun1':n1,'Verb':v,'Noun2':n2}
            dis_list.append(dis_dict)

            # counting the number of sentences containing the verb
            verb = sent.split()[1]
            if verb in verb_dict:
                verb_dict[verb]+=1
            else:
                verb_dict[verb]=1

    return pd.DataFrame(dis_list)
    
```

```python
def dataframe3(df_rule):
    prep_dict = dict()
    dis_dict = dict()
    dis_list = []

# iterating over all the sentences
    for i in range(len(df_rule)):

        # sentence containing the output
        sentence = df_rule.loc[i,'sentences']

        # output of the sentence
        output = df_rule.loc[i,'output']

        # iterating over all the outputs from the sentence
        for sent in output:

            # separate subject, verb and object
            n1 = sent.split()[:1]
            p = sent.split()[1]
            n2 = sent.split()[2:]

            # append to list, along with the sentence
            dis_dict = {'Sent':sentence,'Noun1':n1,'Preposition':p,'Noun2':n2}
            dis_list.append(dis_dict)

    

    return pd.DataFrame(dis_list)
```

```python
def dataframe4(df_rule):
    # separate noun, preposition and noun

    prep_dict = dict()
    dis_dict = dict()
    dis_list = []

# iterating over all the sentences
    for i in range(len(df_rule)):

        # sentence containing the output
        sentence = df_rule.loc[i,'sentences']
        # output of the sentence
        output = df_rule.loc[i,'output']

        # iterating over all the outputs from the sentence
        for sent in output:

            # separate subject, verb and object
            n1 = sent.split()[0]
            p = sent.split()[1]
            n2 = sent.split()[2:]
            
            # append to list, along with the sentence
            dis_dict = {'sentence':sentence,'Noun1':n1.replace('_',' '),'Is/Have':p,'Noun2':n2}
            dis_list.append(dis_dict)

            # counting the number of sentences containing the verb
            prep = sent.split()[1]
            if prep in prep_dict:
                prep_dict[prep]+=1
            else:
                prep_dict[prep]=1

    return pd.DataFrame(dis_list)
```

```python
def rule(rule, text):
    
    if rule == 'rule1':
        df_rule_bok = output(rule1, text)
        df1  = dataframe1(df_rule_bok)
        return df1
    
    elif rule == 'rule3':
        df_rule_bok = output(rule3, text)
        df1  = dataframe3(df_rule_bok)
        return df1
    
    elif rule == 'rule4':
        df_rule_bok = output(rule4, text)
        df1 = dataframe4(df_rule_bok)
        return df1
    
    else :
        print('Error pleas try rule1 or rule3 or rule4')
```

```python
df = joblib.load('df')
```

```python
def make_compounds(compound,text):
    for o in range(len(compound)):
        
        for k in compound[o]:
            mot = k[0]
            deb = k[1]
            fin = k[2]
            text[o] = text[o].replace(text[o][deb:fin],mot.replace(' ','_'))
```

```python
make_compounds(df['compounds'], df['text'])
```

```python
df_rule1 = rule('rule1', df['text'])

df_rule3 = rule('rule3', df['text']) 

df_rule4 = rule('rule4', df['text'])
```
